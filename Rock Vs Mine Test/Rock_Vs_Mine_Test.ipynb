```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Cleaning**

```python
sonar_data=pd.read_csv("/content/sonar.all-data.csv",header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()  #R means Rock & M means Mine
```

```python
sonar_data.groupby(60).mean()
```

Seperating All Data And Labels


```python
X=sonar_data.drop(columns=60,axis=1)
```

```python
Y=sonar_data[60]
```

```python
print(Y)
print()
print(X)
```

Training & Test Data


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.15,stratify=Y,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X.shape)
```

Model Training


```python
Model=LogisticRegression()
```

```python
Model.fit(X_train,Y_train)
```

Finding Accuracy

```python
X_train_predict=Model.predict(X_train)
X_train_accu=accuracy_score(X_train_predict,Y_train)
```

```python
print("Accuracy of Train data is",X_train_accu*100,"%")
```

```python
X_test_predict=Model.predict(X_test)
X_test_accu=accuracy_score(X_test_predict,Y_test)
```

```python
print("Accuracy of Test Data is ",X_test_accu*100,"%")
```
